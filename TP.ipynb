```python
import csv
import graphviz as gv
```

```python
Calles1=[]
with open("LibroCallesTp.csv") as archivo:
  reader=csv.reader(archivo)
  for row in reader:
    Calles1.append(row)
```

```python
Calles1
```

```python
def DibujarGrafo(Grafo, Dirige=False, Camino=[]):
  Grafico = gv.Digraph("felicidad") if Dirige else gv.Graph("alegria")
  n = len(Grafo)
  added = set()
  for Columna, Fila in enumerate(Camino):
    if Fila != -1:
      Grafico.edge(str(Fila), str(Columna), dir="forward", penwidth="2", color="orange")
      added.add(f"{Fila},{Columna}")
      added.add(f"{Columna},{Fila}")
  for Fila in range(n):
    for Columna in Grafo[Fila]:
      if not Dirige and not f"{Fila},{Columna}" in added:  
        added.add(f"{Fila},{Columna}")
        added.add(f"{Columna},{Fila}")
        Grafico.edge(str(Fila), str(Columna))
      elif Dirige:
        Grafico.edge(str(Fila), str(Columna))
  return Grafico

DibujarGrafo(Calles1)
```
